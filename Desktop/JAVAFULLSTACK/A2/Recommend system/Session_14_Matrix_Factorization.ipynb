# Matrix Factorization

Provided rating matrix R of size [number of users, number of movies]. Rating scores range from 1 to 5. Rating of 0 indicates that user has not seen the movive


We are going to find user - feature matrix P of size [number of users, number of features] and item - feature matrix Q of size [number of items, number of features] such that 
$$ R = P.Q^{T} $$

P and Q are initialized randomly, and can be optimized using Gradient Descent

Predicted value and error for each user-movie pair is calculated with:
$$ \hat{r_{ij}} = p_i^{T}q_j = \sum_{u=1}^{k}p_{iu}q_{uj} $$
$$ e_{ij} = (r_{ij} - \hat{r}_{ij})^2 $$

Find gradient by taking differential of error for each feature
$$ \frac{\partial}{\partial p_{iu}}e_{ij}^2 = -2e_{ij}q_{ju} $$
$$ \frac{\partial}{\partial q_{ju}}e_{ij}^2 = -2e_{ij}p_{iu} $$
